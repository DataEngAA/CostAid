### Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select numerical columns for PCA
pca_data = merged_df[['Government Hospitals', 'Private Hospitals', 'Total Hospitals', 'Average Cost in Private Hospitals (₹)', 'Average Cost in Government Hospitals (₹)']]

# Standardize the data
scaler = StandardScaler()
pca_data_scaled = scaler.fit_transform(pca_data)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(pca_data_scaled)

# Create a DataFrame with principal components
pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])

# Plot PCA
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['Principal Component 1'], pca_df['Principal Component 2'])
plt.title('PCA of Hospital and Cost Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
statewise_hospitals = pd.read_excel('C:/Users/user/Documents/Internship/Project Data/State wise - H.xlsx')
average_costs = pd.read_excel('C:/Users/user/Documents/Internship/Project Data/Average Cost.xlsx')

# Merge the datasets on the 'State/Union Territory' column
merged_data = pd.merge(statewise_hospitals, average_costs, on='State/Union Territory')

# Set the figure size
plt.figure(figsize=(14, 8))

# Plot the number of hospitals as a bar chart
sns.barplot(x='State/Union Territory', y='Total Hospitals', data=merged_data, color='b', label='Total Hospitals')

# Create a secondary y-axis to plot the average cost
ax2 = plt.twinx()

# Plot the average cost in private hospitals as a line chart
sns.lineplot(x='State/Union Territory', y='Average Cost in Private Hospitals (₹)', data=merged_data, ax=ax2, color='r', marker='o', label='Average Cost in Private Hospitals (₹)')

# Plot the average cost in government hospitals as a line chart
sns.lineplot(x='State/Union Territory', y='Average Cost in Government Hospitals (₹)', data=merged_data, ax=ax2, color='g', marker='o', label='Average Cost in Government Hospitals (₹)')

# Set the labels and title
plt.title('Total Hospitals and Average Costs in Each State')
ax2.set_ylabel('Average Cost (₹)')
plt.xlabel('State/Union Territory')
plt.ylabel('Total Hospitals')
plt.xticks(rotation=90)

# Add legends
plt.legend(loc='upper left')
ax2.legend(loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import re

# Load the dataset
insurance_cv = pd.read_excel('C:/Users/user/Documents/Internship/Project Data/Insurance Coverage and Economic Indicators.xlsx')

# Function to extract numeric value from percentage string
def extract_numeric(value):
    if isinstance(value, str):
        # Use regular expression to find numbers in the string
        match = re.search(r'[\d.]+', value)
        if match:
            return float(match.group(0))
    # Return NaN for non-string values or if no match is found
    return None

# Apply the function to the 'Value' column and convert to numeric
insurance_cv['Value'] = insurance_cv['Value'].apply(extract_numeric)

# Drop any rows with NaN values in the 'Value' column after conversion
insurance_cv_cleaned = insurance_cv.dropna(subset=['Value'])

# Prepare the data
labels = insurance_cv_cleaned['Indicator']
sizes = insurance_cv_cleaned['Value']

# Plot
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=plt.get_cmap('tab20').colors, startangle=140)
plt.title('Insurance Coverage and Economic Indicators')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
