# Insurance Groups Analysis

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

## Analysing the No. of Policies, People Covered and Gross Premium 

```python
df = pd.read_excel("C:\\Users\\MAHIMA\\Downloads\\Health Insurance data.xlsx", sheet_name = '62', header=[2,3])
df.head()
```

```python
df = df.drop(columns = 'S.No.', level=0)
df.head()
```

```python
#Identifying the columns in the dataframe
for i in range(2):
    print(f"Level {i} :{df.columns.levels[i]}")
```

```python
# Let us rename the scheme names to facilitate easier manipulation
rename_dict={ 
    "No. of policies": "policy_count",
    "No. of Persons Covered ('000s)": "persons_covered",
    "Gross Premium (₹Lakh)": "gross_premium"
}
df = df.rename(columns=rename_dict)
df.head()
```

```python
#Finding the residual unformatted column names
col_names = list(df.columns)
cols_2n3 = list()
for name in col_names:
    cols_2n3.extend([name[1]])
set(cols_2n3)
```

```python
#Removing the sub-level under the insurers column
ins_values = df['Insurers'].values
ins_df = pd.DataFrame(ins_values)
df = df.drop(columns = ('Insurers', 'Unnamed: 1_level_1'))
df.insert(0, "Insurers", ins_df)

df.head()
```

```python
df1 = df.copy()
```

```python
# The timeframe of the data

#Level 0 values
fin_years = set(df.columns.get_level_values(0))
fin_years.discard('Insurers')
fin_years = tuple(fin_years)

fin_years = sorted(fin_years, key= lambda year: int(year[-2:]))
fin_years
```

```python
#Multiplying the columns by the scale value 
x = df1.copy()
policies_df_by_yr = {}
for year in fin_years:
    try:
        x[(year,'persons_covered')] = x[(year,'persons_covered')].apply(lambda x:x*1000)
        print(f"Success=>[({year}, 'persons_covered')]")
        x[(year,'gross_premium')] = x[(year,'gross_premium')].apply(lambda x:x*10**5)
        print(f"Success=>[({year}, 'gross_premium')]")
    except:
        print(f"Failure=>[({year})]")
```

```python
x.head()
```

```python
cleaned_df1 = x.copy()
```

```python
#selecting the rows with total

cleaned_df1 = cleaned_df1.iloc[[5,29,38]]
cleaned_df1
```

```python
categories = [
    "Public Sector Insurers Total", 
    "Private Sector Insurers Total", 
    "Stand-alone Health Insurers Total"
]
```

### Gross Premium

```python
# Extract the data for gross premium
gross_premium_data = cleaned_df1.iloc[[0,1,2], [3, 6, 9, 12, 15, 18]].values
gross_premium_df = pd.DataFrame(gross_premium_data, columns=fin_years, index=categories)
```

```python
gross_premium_df
```

```python
# Plotting
colours = ["yellow", "orange", "red"]

gross_premium_df.T.plot(kind='bar', color = colours, figsize=(8, 5))
plt.title('Gross Premium Over the Years')
plt.xlabel('Years')
plt.ylabel('Gross Premium (in billions)')
plt.xticks(rotation=45)
plt.legend(title='Insurer Category')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```

## Analysis Net Earned Premium and Net Claims Incurred

```python
data = pd.read_excel("C:\\Users\\MAHIMA\\Downloads\\Health Insurance data.xlsx", sheet_name = '66', header=[2,3])
data.head()
```

```python
data = data.drop(columns = 'S.No.', level=0)
data.head()
```

```python
#Identifying the columns in the dataframe
for i in range(2):
    print(f"Level {i} :{data.columns.levels[i]}")
```

```python
# Let us rename the scheme names to facilitate easier manipulation
rename_dict={ 
    "Claims Incurred (Net)": "net_claims_incurred",
    "Incurred Claims Ratio": "incc_claim_ratio(%)",
    "Net Earned Premium": "net_earned_premium"
}
data = data.rename(columns=rename_dict)
data.head()
```

```python
#Removing the sub-level under the insurers column
ins_values = data['Insurers'].values
ins_df = pd.DataFrame(ins_values)
data = data.drop(columns = ('Insurers', 'Unnamed: 1_level_1'))
data.insert(0, "Insurers", ins_df)

data.head()
```

```python
# The timeframe of the data

#Level 0 values
fin_years = set(data.columns.get_level_values(0))
fin_years.discard('Insurers')
fin_years = tuple(fin_years)

fin_years = sorted(fin_years, key= lambda year: int(year[-2:]))
fin_years
```

```python
data1 = data.copy()
data1
```

```python
#selecting the rows with total
data1 = data1.iloc[[5,29,38,39]]
data1
```

### Net Earned Premium

```python
net_earned_premium = data1.iloc[[0,1,2], [1, 4, 7, 10, 13, 16]].values.astype(float)
```

```python
# Plotting the lines for each category
plt.figure(figsize=(8, 5))
for i, category in enumerate(categories):
    plt.plot(fin_years, net_earned_premium[i], marker='o', label=category)

# Customize the plot
plt.title('Net Premium Earned Over the Years by Insurer Groups')
plt.xlabel('Years')
plt.ylabel('Net Premium')
plt.xticks(rotation=45)
plt.legend(title='Insurer Groups')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Net Claims Incurred

```python
net_claims_incurred = data1.iloc[[0,1,2], [2, 5, 8, 11, 14, 17]].values.astype(float)
```

```python
# Plotting the lines for each category
plt.figure(figsize=(8, 5))
for i, category in enumerate(categories):
    plt.plot(fin_years, net_claims_incurred[i], marker='o', label=category)

# Customize the plot
plt.title('Net Claims Incurred Over the Years by Insurer Groups')
plt.xlabel('Years')
plt.ylabel('Net Claims Incurred')
plt.xticks(rotation=45)
plt.legend(title='Insurer Groups')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Trend of Grand Total

```python
net_earned_premium = data1.iloc[3, [1, 4, 7, 10, 13, 16]].values.astype(float)
net_claims_incurred = data1.iloc[3, [2, 5, 8, 11, 14, 17]].values.astype(float)
```

```python
# Plotting the lines
plt.figure(figsize=(8, 5))
plt.plot(fin_years, net_earned_premium, marker='o', label='Net Earned Premium', color='blue')
plt.plot(fin_years, net_claims_incurred, marker='o', label='Net Claims Incurred', color='green')

# Customize the plot
plt.title('Net Earned Premium Vs Claims Incurred Over the Years')
plt.xlabel('Years')
plt.ylabel('Grand Total')
plt.xticks(rotation=45)
plt.legend(title='Metrics')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
